```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df=sklearn.datasets.load_boston()
df
```

```python
# Loading dataset to pandas dataframe
df1=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
df1
```

```python
df1['prices']=df.target
```

```python
df1
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
c=df1.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(c,cbar=True)
```

```python
x=df1.drop(['prices'],axis=1)
y=df1['prices']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
# Prediction on training data
pred=model.predict(x_train)
```

```python
# R squared error on training data
score_1=metrics.r2_score(y_train,pred)
print('R squared error on training data : ',score_1)

#Mean absolute error on training data
score_2=metrics.mean_absolute_error(y_train,pred)
print('Mean absolute error on training data :',score_2)
```

```python
# Visualization 
plt.scatter(y_train,pred)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual Prices VS Predicted Prices')
plt.show()
```

```python
test_pred=model.predict(x_test)
```

```python
# R squared error
score_1=metrics.r2_score(y_test,test_pred)
print('R squared error on testing data : ',score_1)

#Mean absolute error
score_2=metrics.mean_absolute_error(y_test,test_pred)
print('Mean absolute error on testing data :',score_2)
```
